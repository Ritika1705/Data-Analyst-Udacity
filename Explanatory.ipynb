## UNIVARIATE EXPLORATION

```python
b_col = sns.color_palette()[0]
sns.countplot(data = df, x = 'Term',color=b_col)
plt.title('Count of Term');
```

> From the above graph, it is clearly seen that, The maximum number of people issue loan for a period of 36 months and minimum number of people issue for a period of 12 months.

```python
b_col = sns.color_palette()[0]
cat_order = df.Term.value_counts().index
sns.countplot(data = df, x = 'Term',color=b_col,order= cat_order);
```

```python
b_col = sns.color_palette()[0]
sns.countplot(data = df, x = 'LoanStatus',color=b_col)
plt.xticks(rotation=90)
plt.title('Count of Loanstatus');
```

> In the graphs of Loan Status, we can find that, Current Loan status has the maximum count followed by the completed Loan status with Chargeoff, Defaulted and Past Due, following next.

```python
n_p = df.shape[0]
max_cnt = df['LoanStatus'].value_counts().max()
max_prop = max_cnt / n_p
tick_props = np.arange(0, max_prop, 0.05)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
base_color = sns.color_palette()[0]
sns.countplot(data = df, x = 'LoanStatus', color = base_color)
plt.yticks(tick_props * n_p, tick_names)
plt.ylabel('proportion')
plt.xticks(rotation=90);
```

```python
b_col = sns.color_palette()[2]
sns.countplot(data = df, x = 'CreditGrade',color=b_col)
plt.xticks(rotation=90);
```

```python
bin_edges = np.arange(0, df['BorrowerRate'].max()+1, 1)
plt.hist(data = df, x = 'BorrowerRate')
plt.title('BorrowRate Distribution')
plt.xlabel('BorrowerRate')
plt.ylabel('Count');
```

> We can clearly see the distribution of borrowrate count, int the graphs above and below. And we can also see that, the mean of the borrowrate is around 1.9 or 2

```python
sns.distplot(df['BorrowerRate']);
```

## MULTIVARIATE EXPLORATION

```python
g = sns.FacetGrid(data = df, hue = 'IncomeRange', height = 5)
g.map(plt.scatter,  'BorrowerRate','EstimatedReturn')
g.add_legend()
plt.title('BorrowerRate vs EstimatedReturn based on Income');
```

```python
plt.scatter(data = df, x = 'BorrowerRate', y = 'EstimatedLoss', c = 'ProsperScore')
plt.colorbar()
plt.title('BorrowerRate vs EstimatedLoss');
```

#### In the above multivariate exploration graphs, we can see the distribution of BorrowerRate vs. EstimatedReturn based on income.
#### We can also see the relationship between BorrowerRate and EstimatedLoss based on ProsperScore
