**The code snippet below demonstrates the use of function for code length reduction**

```python
# Use of a function

def countplot( a , b):
    sns.countplot(data = a, x = b,color=b_col)
```

### WHAT IS THE TERM RATE IN WHICH PEOPLE ARE MOST INTERESTED IN INVESTING ?

```python
b_col = sns.color_palette()[0]
countplot(df,'LoanStatus')
plt.xticks(rotation=90)
plt.title('Count of Loanstatus')
```

#### WHAT IS THE OBSERVATION FROM THE ABOVE GRAPH ?

> On examining the loan Status, we can see that, current loan status is the maximum followed by the completed loan status.

```python
n_p = df.shape[0]
max_cnt = df['LoanStatus'].value_counts().max()
max_prop = max_cnt / n_p
tick_props = np.arange(0, max_prop, 0.05)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
base_color = sns.color_palette()[0]
sns.countplot(data = df, x = 'LoanStatus', color = base_color)
plt.yticks(tick_props * n_p, tick_names)
plt.ylabel('proportion')
plt.xticks(rotation=90);
```

## BIVARIATE EXPLORATION

```python
plt.scatter(data = df, x = 'BorrowerRate', y = 'LenderYield')
plt.title('BorrowerRate vs LenderYield')
```

**USE OF FUNCTION TO CREATE REGPLOT**

```python
def rplt(a , b, c):
    sns.regplot(data = a, x = b, y = c, scatter_kws = {'alpha':1/10})
```

```python
rplt(df,'BorrowerRate','LenderYield') # CALLING FUNCTION RPLT.
plt.title('BorrowerRate vs LenderYield')
```

```python
rplt(df,'BorrowerAPR','LenderYield')
plt.title('BorrowerAPR vs LenderYield');
```

#### After plotting the scatter plots of BoorowerRate vs Lender yield and BorrowerAPR vs LenderYield, we find that BorrowerAPR is a more effective and clear measure.

```python
sns.regplot(data = df, x = 'BorrowerAPR', y = 'BorrowerRate', scatter_kws = {'alpha' : 1/10})
```

#### WHAT IS THE RELATIONSHIP BETWEEN LOANSTATUS AND LENDERYIELD?

**The above violin plots represent the relationship between Loan Status vs. Lender Yeild. The white dot on every plot represents the median, whereas the area represents interquartile range.**

**The above plot represents the relationship between CreditGrade and LenderYield. The shapes where the edges are extremely skinny and thick at the centre represents that LenderYield is concentrated around the median.**
